```python
# import libs
from imageai.Detection import ObjectDetection
import os
```

```python
# make detector object for applying model and its path
detector = ObjectDetection()
# set model
detector.setModelTypeAsYOLOv3()
# path of path from local machine
detector.setModelPath( os.path.join("/home/samruddhak/Documents" , "yolov3.pt"))
# load the model
detector.loadModel()
```

```python
# target image path and saving detected image path 
detections = detector.detectObjectsFromImage(input_image=os.path.join("/home/samruddhak/Documents"  , "traffic.jpeg"), output_image_path=os.path.join("/home/samruddhak/Documents"  , "object_detected.jpg"), minimum_percentage_probability=30)
```

```python
# get the lables and its accuracy 
for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```
